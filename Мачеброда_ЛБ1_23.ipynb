```python
#Створіть одновимірний масив (вектор) зі значеннями квадратних коренів з кожного натурального числа, починаючи з 1, та виведіть його значення.
import math

# Встановіть кількість натуральних чисел, для яких потрібно знайти квадратний корінь
n = 10  # Наприклад, для перших 10 чисел

# Створюємо список з квадратних коренів
square_roots = [math.sqrt(i) for i in range(1, n + 1)]

# Виводимо значення
print("Масив квадратних коренів:")
print(square_roots)
```

```python
#Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його значеннями коренів 4 ступеня перших дійсних чисел
import numpy as np

# Розмір матриці
rows, cols = 3, 3

# Кількість елементів у матриці
num_elements = rows * cols

# Створюємо матрицю розміром 3x3, заповнену коренями четвертого ступеня
matrix = np.zeros((rows, cols))

# Заповнюємо матрицю значеннями коренів четвертого ступеня
for i in range(rows):
    for j in range(cols):
        number = i * cols + j + 1  # Дійсні числа починаються з 1
        matrix[i, j] = number ** (1/4)

# Виводимо матрицю
print("Матриця коренів четвертого ступеня:")
print(matrix)
```

```python
#Створіть матрицю розміром 2x2, заповніть її випадковими цілими числами в діапазоні від 1 до 15 та знайдіть її обернену матрицю.
import numpy as np

# Розмір матриці
size = 2

# Генеруємо матрицю розміром 2x2 з випадковими цілими числами від 1 до 15
matrix = np.random.randint(1, 16, size=(size, size))

# Виводимо матрицю
print("Матриця:")
print(matrix)

# Перевіряємо, чи матриця є оберненою (перевіряючи детермінант)
det = np.linalg.det(matrix)
if det == 0:
    print("\nМатриця є виродженою (детермінант дорівнює 0). Обернена матриця не існує.")
else:
    # Знаходимо обернену матрицю
    inverse_matrix = np.linalg.inv(matrix)

    # Виводимо обернену матрицю
    print("\nОбернена матриця:")
    print(inverse_matrix)
```

```python
#Створіть матрицю розміром 3x4 та вектор розміром 4x1, заповніть матрицю випадковими цілими числами в діапазоні від 1 до 10, а вектор - випадковими цілими числами в діапазоні від 1 до 10, і знайдіть добуток матриці на вектор.
import numpy as np

# Розміри матриці та вектора
matrix_rows, matrix_cols = 3, 4
vector_size = 4

# Генеруємо матрицю розміром 3x4 з випадковими цілими числами від 1 до 10
matrix = np.random.randint(1, 11, size=(matrix_rows, matrix_cols))

# Генеруємо вектор розміром 4x1 з випадковими цілими числами від 1 до 10
vector = np.random.randint(1, 11, size=(vector_size, 1))

# Виводимо матрицю та вектор
print("Матриця (розмір 3x4):")
print(matrix)
print("\nВектор (розмір 4x1):")
print(vector)

# Знаходимо добуток матриці на вектор
product = np.dot(matrix, vector)

# Виводимо результат
print("\nДобуток матриці на вектор:")
print(product)
```

```python
#Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від 1 до 5 та знайдіть добуток всіх елементів першої матриці
import numpy as np
from functools import reduce
import operator

# Розмір матриць
size = 3

# Генеруємо дві матриці розміром 3x3 з випадковими цілими числами від 1 до 5
matrix1 = np.random.randint(1, 6, size=(size, size))
matrix2 = np.random.randint(1, 6, size=(size, size))

# Виводимо обидві матриці
print("Матриця 1 (розмір 3x3):")
print(matrix1)
print("\nМатриця 2 (розмір 3x3):")
print(matrix2)

# Знаходимо добуток всіх елементів першої матриці
product_of_elements = reduce(operator.mul, matrix1.flatten(), 1)

# Виводимо результат
print("\nДобуток всіх елементів першої матриці:")
print(product_of_elements)
```

```python
#Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від -3 до 3 та знайдіть суму елементів на головній діагоналі.
import numpy as np

# Розмір матриці
size = 3

# Генеруємо матрицю розміром 3x3 з випадковими дійсними числами від -3 до 3
matrix = np.random.uniform(-3, 3, size=(size, size))

# Виводимо матрицю
print("Матриця (розмір 3x3):")
print(matrix)

# Знаходимо суму елементів на головній діагоналі
diagonal_sum = np.trace(matrix)

# Виводимо результат
print("\nСума елементів на головній діагоналі:")
print(diagonal_sum)
```

```python
#Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від -0.5 до 0.5 та знайдіть вектор-стовпчик, що містить суму елементів кожного стовпця матриці.
import numpy as np

# Розмір матриці
size = 3

# Генеруємо матрицю розміром 3x3 з випадковими дійсними числами від -0.5 до 0.5
matrix = np.random.uniform(-0.5, 0.5, size=(size, size))

# Виводимо матрицю
print("Матриця (розмір 3x3):")
print(matrix)

# Знаходимо суму елементів кожного стовпця
column_sums = np.sum(matrix, axis=0)

# Перетворюємо результат на вектор-стовпчик
column_sums_vector = column_sums.reshape(-1, 1)

# Виводимо вектор-стовпчик
print("\nВектор-стовпчик, що містить суму елементів кожного стовпця:")
print(column_sums_vector)
```
